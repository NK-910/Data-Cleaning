```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('all_bikez_curated.csv')
```

```python
df.isna().sum()/len(df)*100
```

```python
df.drop(columns=['Fuel control','Dry weight (kg)'],inplace=True)
```

```python
df.dropna(subset=['Engine cylinder','Engine stroke','Model'],inplace= True)
```

```python
df.isna().sum()/len(df)*100
```

```python
sns.histplot(df['Displacement (ccm)'], kde=True)
plt.show()
```

```python
df['Displacement (ccm)'].fillna(df['Displacement (ccm)'].median(), inplace=True)
```

```python
sns.histplot(df['Power (hp)'], kde=True)
plt.show()
```

```python
df['Power (hp)'].fillna(df['Power (hp)'].median(), inplace=True)
```

```python
sns.histplot(df['Gearbox'], kde=True)
plt.show()
```

```python
df['Gearbox'].fillna(df['Gearbox'].mode(), inplace=True)
```

```python
sns.histplot(df['Fuel capacity (lts)'], kde=True)
plt.show()
```

```python
df['Fuel capacity (lts)'].fillna(df['Fuel capacity (lts)'].median(), inplace=True)
```

```python
sns.histplot(df['Fuel system'], kde=True)
plt.show()
```

```python
df['Cooling system'].fillna(df['Cooling system'].mode(), inplace=True)
```

```python
df['Transmission type'].fillna(df['Transmission type'].mode(), inplace=True)
```

```python
sns.histplot(df['Wheelbase (mm)'], kde=True)
plt.show()
```

```python
df['Wheelbase (mm)'].fillna(df['Wheelbase (mm)'].median(), inplace=True)
```

```python
sns.histplot(df['Seat height (mm)'], kde=True)
plt.show()
```

```python
df['Seat height (mm)'].fillna(df['Seat height (mm)'].median(), inplace=True)
```
